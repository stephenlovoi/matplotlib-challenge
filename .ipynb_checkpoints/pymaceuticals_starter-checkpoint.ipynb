## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_study_df = pd.merge(mouse_metadata, study_results, 
                          how="outer", on="Mouse ID")

# Display the data table for preview
mouse_study_df
```

```python
# Checking the number of mice.
mice_count = len(mouse_study_df["Mouse ID"].value_counts())
print(f"Number of Mice: {mice_count}")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_study_df["Mouse ID"].value_counts()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
g989_mouse = mouse_study_df.loc[mouse_study_df["Mouse ID"] == "g989", :]

g989_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_df = mouse_study_df[mouse_study_df['Mouse ID'] != 'g989']
clean_mouse_df
```

```python
#Drop rows where 'Mouse ID' only appears once
```

```python
# Checking the number of mice in the clean DataFrame.
new_mice_count = len(clean_mouse_df["Mouse ID"].value_counts())
print(f"New Number of Mice: {new_mice_count}")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
grouped_regimen_df = clean_mouse_df.groupby(["Drug Regimen"])

tumor_volume = grouped_regimen_df["Tumor Volume (mm3)"]

regimen_mean = tumor_volume.mean()

regimen_median = tumor_volume.median()

regimen_variance = tumor_volume.var()

regimen_std = tumor_volume.std()

regimen_sem = tumor_volume.sem()

# Assemble the resulting series into a single summary dataframe.
summary_stats_df = pd.DataFrame({"Mean Tumor Volume (mm3)": regimen_mean, "Median Tumor Volume (mm3)": regimen_median, 
                                "Variance of Tumor Volume (mm3)": regimen_variance, "Std. Dev.of Tumor Volume (mm3)": regimen_std, 
                                "SEM of Tumor Volume (mm3)": regimen_sem})
summary_stats_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

summary_stats_2_df = clean_mouse_df.groupby('Drug Regimen').agg(['mean', 'median', 'var', 'std', 'sem'])["Tumor Volume (mm3)"]
summary_stats_2_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pandas.
drug_regimen = pd.DataFrame(clean_mouse_df.groupby(["Drug Regimen"]).count()["Mouse ID"])
drug_regimen.plot(kind = 'bar', color = 'b', figsize = (10, 3), legend = False)


plt.xlabel("Drug Regimen")
plt.ylabel("Count")
plt.show()
```

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pyplot.
x_axis = np.arange(len(drug_regimen))
y_axis = drug_regimen["Mouse ID"]

plt.bar(x_axis, y_axis, color = 'r', alpha = 0.7, align = 'center')
bar_tick_locations = [value for value in x_axis]
plt.xticks(bar_tick_locations, drug_regimen.index, rotation=90)
plt.show()
```
